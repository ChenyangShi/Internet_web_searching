```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib2
import warnings
warnings.filterwarnings("ignore")

#Canada 
web = 'https://www.sports-reference.com/olympics/countries/CHN/summer/2016/'
#USA
#
req = urllib2.Request(web)#request to visit web
page = urllib2.urlopen(req)#open this url
soup = BeautifulSoup(page, 'lxml')##将网页源码构造成BeautifulSoup对象，方便操作
table = soup.find("div",{"id" : "div_athletes"})#找到这个标签
print soup.prettify()
```

```python
A=[]

for a in table.find_all('a', href=True):
    A.append(a['href'])
    
A
```

```python
#加上完整的url链接
URLfull = []
for i in A:
    URLfull.append('http://www.sports-reference.com' + i)
```
