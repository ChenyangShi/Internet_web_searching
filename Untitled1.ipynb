```python
import re
removeImg = re.compile('<img.*?>| {7}|')#去除img标签,7位长空格
removeAddr = re.compile('<a.*?>|</a>')#删除超链接标签
replaceLine = re.compile('<tr>|<div>|</div>|</p>')#把换行的标签换为\n
replaceFont = re.compile('<font.*?>|</font>')#去除font标签,7位长空格
replaceBr = re.compile('<br.*?>|</br>')
replaceH = re.compile('<h3.*?>|</h3>')
replaceStrong = re.compile('<strong.*?>|</strong>')

def Replace(x):
    x = re.sub(removeImg, "", x)
    x = re.sub(removeAddr, "", x)
    x = re.sub(replaceLine, "", x)
    x = re.sub(replaceFont, "", x)
    x = re.sub(replaceBr, "", x)
    x = re.sub(replaceH, "", x)
    x = re.sub(replaceStrong, "", x)
    return x.strip();

#print str(mainTags[0])
a = Replace(str(mainTags[0]))
print a
#a = re.sub(replaceFont, "", str(mainTags[0])).strip();
#print a
```
